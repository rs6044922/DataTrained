## Treating the outlier

```python
#removing the outlier by Z score method
from scipy.stats import zscore
df_new=df
z=np.abs(zscore(df_new))
df_new=df_new[(z<3).all(axis=1)]
print("Before removing the outlier ",df.shape)
print('After removing the outlier ',df_new.shape)
p=100-(len(df_new.index)/len(df.index)*100)
print(f'We are lossing {p} % of data.')
```

## Data set is very little record so we are not removing the outlier

## Scaling the data

```python
#Creating the dependent and independent variable 
x=df.drop('W',axis=1)
y=df['W']
```

```python
#scaling the data using the standard data 
from sklearn.preprocessing import MinMaxScaler
MS=MinMaxScaler()
for i in x:
  x[i]=MS.fit_transform(x[[i]])
```

Scaling the data by Min and Max Scalar

## Treating the skewness

```python
#checking the skewness of each column
x.skew()
```

```python
#applying power transform 
from sklearn.preprocessing import PowerTransformer
PT=PowerTransformer()
x['R']=PT.fit_transform(x[['R']])
x['H']=PT.fit_transform(x[['H']])
x['CG']=PT.fit_transform(x[['CG']])
x['SV']=PT.fit_transform(x[['SV']])
x['E']=PT.fit_transform(x[['E']])
```

```python
#checking the skewness after power transform
x.skew()
```

We have removed the skewness and all the column is with in the -0.5 to 0.5

# Building the model 
## Selecting the best random state

```python
#importing the library for model building
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#selecting the best random state
from sklearn.linear_model import LinearRegression

mA=0
mRS=0
for i in range(1,100):
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
  mod=LinearRegression()
  mod.fit(x_train,y_train)
  pred=mod.predict(x_test)
  r2=r2_score(y_test,pred)
  if r2>mA:
    mA=r2
    mRS=i
print('Best accuracy is ',mA,' on random state ',mRS)
```

## Linear Regression 

```python
#building the model using LinearRegression, RandomForestRegressor, DecissionTreeRegressor, svrkernal, Lasso
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
#spliting the data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=5)
```

```python
#building model 
model={LinearRegression():"LinearRegression", DecisionTreeRegressor():'DecisionTreeRegressor', RandomForestRegressor():'RandomForestRegressor', Lasso():'Lasso'}
r2score=[]
crossvalidation=[]
for key in model:
  print("===== ",model.get(key),' model ===== ')
  key.fit(x_train,y_train)
  pred=key.predict(x_test)
  r2=r2_score(y_test,pred)
  print("R2 Score / Accuracy Score is ",r2)
  r2score.append(r2)
  c=cross_val_score(key,x,y,cv=5)
  print('Cross validation score is',(np.abs(c)).mean(),'\n')
  crossvalidation.append()
  mse=mean_squared_error(y_test,pred)
  print('Mean Squared Error',mse)
  mae=mean_absolute_error(y_test,pred)
  print('Mean Absolute Error',mae)
  print('\n\n\n')
```

```python
#selecting the best model
Modelselection=pd.DataFrame({})
Modelselection['Model_Name']=['LinearRegression','DecisionTreeRegressor', 'RandomForestRegressor','Lasso']
Modelselection['R2 Score']=r2score
Modelselection['Cross Validation']=crossvalidation
Modelselection['Diff of R2 and Cross_Val']=Modelselection['R2 Score']-Modelselection['Cross Validation']
Modelselection
```

From the above tabel we can see that the linear regression has high accuracy of 94% with minimum deviation of 7%
So we will select Linear regression as our dest model and we will carry out Hyper parameter testing for that model

# Hyper parameter tunning 

```python
#carring out hyper parameter tunning 
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

p1={'normalize':['True'],'fit_intercept':['True'],'n_jobs':[4]}
p2={'n_jobs':[5],'copy_X':['True'],}
p=(p1,p2)

for i in p:
  l_r=LinearRegression()
  clf=GridSearchCV(l_r,i)
  clf.fit(x_test,y_test)
  print(clf.best_params_)
```

We have carried out hyper parameter testing with 2 set one contains 3 parameter and another contains 2 parameter 
and we have got best parameter for both 2 sets
Now i will build the 2 model using both best parameters

MODEL I 

```python
#tuning on model_1
model_1=LinearRegression(fit_intercept=True, n_jobs=4, normalize=True)
model_1.fit(x_train,y_train)
pred=model_1.predict(x_test)
r2=r2_score(y_test,pred)
print("R2 Score / Accuracy Score is ",r2)
c=cross_val_score(model_1,x,y,cv=5)
print('Cross validation score is',(np.abs(c)).mean(),'\n')
mse=mean_squared_error(y_test,pred)
print('Mean Squared Error',mse)
mae=mean_absolute_error(y_test,pred)
print('Mean Absolute Error',mae)
```

MODEL II

```python
#tuning on model_2
model_2=LinearRegression(copy_X=True, n_jobs=5)
model_2.fit(x_train,y_train)
pred=model_2.predict(x_test)
r2=r2_score(y_test,pred)
print("R2 Score / Accuracy Score is ",r2)
c=cross_val_score(model_2,x,y,cv=5)
print('Cross validation score is',(np.abs(c)).mean(),'\n')
mse=mean_squared_error(y_test,pred)
print('Mean Squared Error',mse)
mae=mean_absolute_error(y_test,pred)
print('Mean Absolute Error',mae)
```

We can observe that there is no change after hyper parameter testing, Now lets stick on to base LinearRegression model that is as follows

```python
#Final_model
Final_model=LinearRegression()
print("===== Final Linear_Regression Model ===== ")
Final_model.fit(x_train,y_train)
pred=Final_model.predict(x_test)
r2=r2_score(y_test,pred)
print("R2 Score / Accuracy Score is ",r2)
c=cross_val_score(Final_model,x,y,cv=5)
print('Cross validation score is',(np.abs(c)).mean(),'\n')
mse=mean_squared_error(y_test,pred)
print('Mean Squared Error',mse)
mae=mean_absolute_error(y_test,pred)
print('Mean Absolute Error',mae)
```
